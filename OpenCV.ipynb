```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/images/aurora_image.jpg')
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# There are three layers in which we see images
# by default image is read in Blue Green Red layers
```

```python
plt.style.use('seaborn')
plt.imshow(img)
plt.show()
```

```python
plt.style.use('seaborn')
# plt.imshow(img)
plt.imshow(rgb_image)
plt.axis("off")
plt.show()
```

```python
# Cropping a part of image
crp_image = rgb_image[4000:5000, 2000:3000 ,:]
plt.imshow(crp_image)
plt.show()
```

```python
crp_image.shape
```

```python
total_pixels = crp_image.shape[0]*crp_image.shape[1]
# Each pixel will have 2 integer values
total_pixels
```

```python
# Resizing an image
crp_image = cv2.resize(crp_image, (100 ,100))
plt.imshow(crp_image)
plt.axis("off")
plt.plot()
plt.title("Man"+str(crp_image.shape))
plt.show()
```
