```python
#Q17 Which product categories are most profitable?  Draw   a  bar  plot  and find   the  answer?
```

```python
import matplotlib.pyplot as plt
x_axis = df.groupby('Category').agg({'Profit':sum})['Profit']
y_axis = df.Category.unique()
plt.xlabel('Category')
plt.ylabel('Profit')
plt.bar(y_axis, x_axis)
plt.show()
```

Technology category is most profitable

```python
# 19 Which region  are the most profitable? draw a   bar  plot and  find answer?
```

```python
import matplotlib.pyplot as plt
x_axis = df.groupby('Region').agg({'Profit':sum})['Profit']
y_axis = df.Region.unique()
plt.xlabel('Region')
plt.ylabel('Profit')
plt.bar(y_axis, x_axis)
plt.show()
```

East region is most profitable
