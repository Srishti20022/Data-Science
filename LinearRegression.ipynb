```python
x_train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/LinearRegresseion/Training Data/Linear_X_Train.csv')
y_train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/LinearRegresseion/Training Data/Linear_Y_Train.csv')

# Converting to arrays
x_train = np.array(x_train)
y_train = np.array(y_train)

# Normalization
u = x_train.mean()
std = x_train.std()
x_train = (x_train-u)/std
plt.title("Time vs Performance graph")
plt.xlabel("Time")
plt.ylabel("Performance")

plt.scatter(x_train, y_train)
plt.show()
```

```python
# Returning predicted y which is y_
def hypothesis(x, theta) :
    y_ = theta[0] + theta[1]*x
    return y_

def gradient(x, y, theta):
    grad = np.zeros((2,))
    size = x.shape[0]
    for i in range (size):
        y_ = hypothesis(x[i], theta)
        grad[0] += (y_-y[i])
        grad[1] += (y_-y[i])*x[i]
    return grad/size

def error(x, y, theta):
    err = 0.0
    n = x.shape[0]
    for i in range(n):
        y_pred = hypothesis(x[i], theta)
        err += (y[i]-y_pred)**2
    return err/n

def gradientDescent(x, y, steps = 200, learningRate = 0.1) :
    theta = np.zeros((2,))
    size = x.shape[0]
    error_list = []
    for i in range(steps):
        grad = gradient(x, y, theta)
        a = error(x, y, theta)
        if i == 0:
            print(a)
        error_list.append(a)
        theta[0] = theta[0] - learningRate*grad[0]
        theta[1] = theta[1] - learningRate*grad[1]
        
        

    return error_list
```

```python
error_list = gradientDescent(x_train, y_train)
print(theta)
# print(error_list)
```
